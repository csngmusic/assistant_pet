```python
from pdf2image import convert_from_path
import cv2
import numpy as np
import pytesseract
from config import *

pytesseract.pytesseract.tesseract_cmd = tess_path

# Предварительная обработка изображения
def preprocess_image(image):
    """Функция для предварительной обработки изображения перед OCR"""
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)  # Перевод в градации серого
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)  # Бинаризация
    return binary

# Распознавание текста
def extract_text(pdf_path):
    images = convert_from_path(pdf_path, dpi=300)
    text = ""
    for img in images:
        img = np.array(img)
        processed_img = preprocess_image(img)
        temp = pytesseract.image_to_string(processed_img, lang="eng+rus+math", config="--oem 3")
        print(temp)
        text += temp
    return text
```

```python
pdf_path = 'files/моносахариды.pdf'

text = extract_text(pdf_path)

text
```
